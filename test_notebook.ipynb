```python
import requests 
  
# api-endpoint 
URL = "http://lookup.dbpedia.org/api/search/KeywordSearch?QueryClass=film&QueryString=toy story&MaxHits=1"
headers = {'Accept':'application/json'} 
r = requests.get(url = URL, headers=headers) 
  
# extracting data in json format 
from urllib.parse import unquote
print(unquote("http://dbpedia.org/sparql/?default-graph-uri=http%3A%2F%2Fdbpedia.org&query=DESCRIBE+%3Chttp%3A%2F%2Fdbpedia.org%2Fresource%2FToy_Story%3E&format=text%2Fcsv&CXML_redir_for_subjs=121&CXML_redir_for_hrefs=&timeout=30000&debug=on&run=+Run+Query+"))

data = r.json() 
# print(data)
r2 = requests.get(url="http://dbpedia.org/sparql/?default-graph-uri=http://dbpedia.org&query=DESCRIBE+<http://dbpedia.org/resource/Toy_Story>&format=text/csv")
triplets = r2.content.decode('utf-8').split('\n')[1:]
[[i.strip('"') for i in triple.split(',')] for triple in triplets]
```
